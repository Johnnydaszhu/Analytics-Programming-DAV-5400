```python
import requests
import pandas as pd
from pprint import pprint
from pandas.io.json import json_normalize
import json

api_123 = open("/Users/johnnyzhu/Downloads/nyt_api.txt", "r").readline()
```

```python
# Top stories on NYTimes.com for specified period
def topstories_data(section):
    # submit our request via the nytimes api
    # note the inclusion of the required API key
    arc=requests.get('https://api.nytimes.com/svc/topstories/v2/'+section+'.json?api-key='+api_123)
    return arc.json();
```

```python
data = topstories_data(input('Use keyword like arts to find the Top stories of New York Times:')) #type a section keyword to find the top stories recently
dl = json_normalize(data['results']) #convert json to pandas dataframe
article_list = pd.DataFrame(dl, columns=['title','byline','geo_facet','published_date']) #select the columns we need
article_list
```
