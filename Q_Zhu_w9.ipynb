## (1) Student_Test_Data

Question: Which student had the highest average test grades?

```python
Student_Test_Data_df = pd.read_csv("https://raw.githubusercontent.com/Johnnydaszhu/DAV-5400/master/Student_Test_Data.csv")
Student_Test_Data_df
```

```python
Student_Test_Data_long_df = Student_Test_Data_df.melt(id_vars = ['Name'], value_vars = ['Test 1', 'Test 2', 'Test 3','Test 4'])
Student_Test_Data_long_df.columns = ['Name', 'Test','Score']
Student_Test_Data_long_df
```

```python
Student_Test_Data_long_df.groupby('Name').mean().idxmax()
```

## (2) Tiding destination_reviews dataset

Question: which class among the six has the highest frequency of reviews by the users.

```python
de_df = pd.read_csv("https://raw.githubusercontent.com/Johnnydaszhu/DAV-5400/master/destination_reviews.csv")
de_df
```

```python
de_long_df = de_df.melt(id_vars = ['User Id'], value_vars = ['Sports', 'Religious', 'Nature','Theatre','Shopping','Picnic'])
de_long_df.columns = ['User Id', 'Destination','Reviews']
de_long_df
```

```python
de_long_df.groupby('Destination').sum().idxmax()
```

## (3) Student_enrollment_data

Question: Compare different institutions’ enrollment by ethnicity.

```python
Student_enrollment_df = pd.read_csv("https://raw.githubusercontent.com/Johnnydaszhu/DAV-5400/master/Student_enrollment_data.csv")
Student_enrollment_df
```

```python
Student_enrollment_long_df = Student_enrollment_df.melt(id_vars = ['Unique ID of the institution', 
                                                                   'Level of student'])
Student_enrollment_long_df.columns = ['Unique ID of the institution', 'Level of student','Ethnicity','Number']
Student_enrollment_long_df
```

```python
Student_enrollment_long_df.groupby('Ethnicity').sum().sort_values(by='Number', ascending=False)
```
